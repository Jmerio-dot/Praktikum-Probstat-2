```python
import pandas as pd

data_agnar = pd.read_clipboard()

# Menampilkan data
print(data_agnar)
```

```python
agnar = data_agnar[data_agnar['Bedrooms'] == 2]

# Menampilkan nama
print(agnar)
```

```python
agnar['Bathrooms'] = pd.to_numeric(agnar['Bathrooms'])

import numpy as np

agnar['Bathrooms'] = agnar['Bathrooms'].apply(lambda x: 'large' if x > 2 else 'small')

# Menampilkan DataFrame setelah modifikasi
print(agnar)
```

```python
import numpy as np

agnar['agnarraka'] = np.where(agnar['Offers'] > 2, 'large', 'small')

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(agnar)
```

```python
# Menambahkan kolom baru 'newvariable' 
agnar['agnarraka'] = agnar['Price'] / agnar['SqFt']

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(agnar)
```

```python
agnar = agnar.drop(columns=['agnarraka'])

# Menampilkan DataFrame 'nama' 
print(agnar)
```

```python
kolom1dan2 = data_agnar.iloc[:, 0:2]

# Menampilkan DataFrame kolom1dan2
print(kolom1dan2)
```

```python
# Memilih kolom 1 dan 2 dari DataFrame data_nama
kolom3dan4 = data_agnar.iloc[:, 2:4]

# Menampilkan DataFrame kolom3dan4
print(kolom3dan4)
```

```python
# Menggabungkan dua DataFrame 
kolom1sd4 = pd.concat([kolom1dan2, kolom3dan4], axis=1)

# Menampilkan DataFrame kolom1sd4
print(kolom1sd4)
```

```python
# Menggabungkan baris dari dua DataFrame
baris1sd3 = data_agnar.iloc[0:3, :]
baris4sd6 = data_agnar.iloc[3:6, :]
baris1sd6 = pd.concat([baris1sd3, baris4sd6])

# Menampilkan DataFrame baris1sd6
print(baris1sd6)
```

```python
data_agnar_sort = data_agnar.sort_values(by='Price')

print(data_agnar_sort)
```
